<h2>ABSTRACT</h2>

This study explores the relationship between the highest educational attainment of the household head and average household health expenditures across regions in the Philippines. With the use of data from the Philippine Statistics Authority’s 2015 Family Income and Expenditure Survey (FIES), this paper will employ an exploratory data analysis with geospatial mapping to highlight regional disparities in education and healthcare spending. <h2>REVIEW OF RELATED LITERATURE</h2>

<h3>I. Education and Health Outcome</h3>

    Extensive research has established a strong link between educational attainment and improved health outcomes. Individuals with higher levels of education generally experience better health, reduced risk of chronic illness, and longer life expectancy. This relationship is found to persist across different cultural and socioeconomic contexts, although with varying magnitudes. The study by Yue et al. (2021) examining middle-aged and older adults in the United States found that having a college degree significantly reduced the likelihood of hospitalization by 8.37% compared to those without a high school diploma. This association persisted even after controlling for income and health status, though it weakened— suggesting that education primarily influences health indirectly by improving overall health conditions. Zajacova and Lawrence (2018) make mention of “the gradient”, as reported in many various studies over the years and replicated in their own study, wherein higher educational attainment is linked to better health outcomes and more years of life. The gradient is seen to hold for all genders and racial groups, with variations in magnitudes.    

    Evidence from China supports these findings, showing that education improves both physical and mental health through various pathways, including higher income, healthier behaviors, and improved psychological well-being. Mental health, in particular, emerged as the strongest mediating factor. The effect was especially pronounced among vulnerable groups, suggesting that education not only improves health outcomes but also reduces health disparities (Long et al., 2020). These findings collectively underscore education’s central role in promoting better and more equitable health.

<h3>II. Education and Health Spending</h3>

    In the global context, while numerous studies have explored the effect of an individual’s educational attainment on the health outcome, only a few have examined the direct relationship between education and health expenditure at the household level. Using the 2004 to 2012 Medical Expenditure Panel Survey, Monheit and Grafova (2018) found in their study that in two-parent families, parents with higher levels of schooling are associated with higher healthcare expenditures, particularly for ambulatory care, dental care, and mental health services. In contrast, the relationship between parental education and household health spending has little evidence in single-mother families. Nonetheless, the results of their study further revealed that an increase in education level in a single-mother family results in an increased spending on prescription drugs and dental care. This indicates that while education can promote greater health investment—likely through increased healthcare awareness, access, and value placed on preventive care—its influence may be limited by structural or financial constraints in single-parent households. Similarly, using the Iranian Household Income and Expenses Survey, Mehrara and Farshchi (2020) reported that there is a positive relationship between the spouse’s education and household healthcare expenditures in an Iranian household. Moreover, Iranian households with spouses who have Ph.D. spend on healthcare 97% more compared to “illiterate” households. 
	
    These findings highlight the differences across family structures, as shown in the weaker association in single-mother households. This highlights a key gap in exploring other socioeconomic and family dynamics factors that might influence household spending on healthcare. Furthermore, there is also a lack of research studies that examine these dynamics in the context of Southeast Asian countries, especially Philippines, where regional disparities in both education and healthcare access are pronounced.   

<h3>III. Other Socioeconomic Determinants of Household Health Spending in ASEAN countries</h3>

    Other than education levels, various socioeconomic factors have been recognized to play an important role in influencing household expenditure. The complex nature of this interplay prompted the identification of other significant variables including income levels, household composition, and geographic location. Using an analysis based on the Philippine Health and Demographic Survey (DHS), a study by Estrada (2023) uncovers micro and macro socioeconomic disparities, investigating how variations in factors like educational attainment, wealth status, household composition, and regional differences can shape an individual's health-related decisions. The results of this study further revealed that individuals in more advantaged socioeconomic positions, thus coming from wealthier households and having attained higher levels of education, were more likely to seek preventive services as they have better access to health information.

    In addition, the presence of elderly or chronically ill members in a household has also been identified as a strong predictor of increased household health spending. This is for the reason that such individuals are expected to typically require sustained medical attention, hence needing prescription medications, routine check-ups, and sometimes long-term care services. In the Philippines, where healthcare access and affordability vary widely by region, this burden is often magnified in rural and underserved areas (Ulep & Dela Cruz, 2013). In consequence, regional inequalities in healthcare infrastructure and education amplify these effects, making the presence of elderly or chronically ill members not just a demographic concern but a critical factor in understanding how households allocate health spending across the country.


<h2>FRAMEWORK AND DATA</h2>

<h3>I. Theoretical Framework</h3>

    This study hinges on Grossman’s (1972) economic insights towards health. In his study, Grossman asserts that an individual's fundamental demand is for the commodity of "good health," which drives their pursuit of acquiring health capital. Thus, the motivation behind expenditure on goods and services that improve health—such as medical procedures and preventive care—is rooted in the demand for "good health" itself, rather than for these specific goods and services. Health is seen as a durable capital stock that produces the output “healthy days”, wherein the decision to invest in health is influenced by the perceived returns from health capital, such as increased productivity and quality of life.

    Grossman also posits that education has a significant effect on both the demand and expenditure for health. He predicts that an increase in education results in an increase in an individual’s efficiency in producing health, which in turn increases the level of health they strive to maintain. More educated people are seen to be able to better manage their health, and reap greater benefits from being healthy— thus demanding a larger optimal stock of health. However, the same straightforward relationship is not recognized between education and health expenditure. Given an inelastic demand for medical services, higher educated individuals may actually spend less— already possessing an improved efficiency in maintaining good health through preventative behavior and informed decision-making.

    A key strength of Grossman’s study is that it explains the influence of variables like age and education through their effect on the cost and productivity of investing in health, rather than through differences in preferences or other factors. 

    Michael Grossman's seminal work on the demand for health (Grossman 1972, 2000) extended the concept of a household production function to the commodity “good health.” In this framework, household members combine inputs of medical care and time to produce health. In Grossman's investment model, such production yields a flow of “healthy days” whose value represents a monetary return over time. Education plays a key role as an “environmental variable” that enhances the efficiency with which resource are used in health production, thereby increasing the marginal product of inputs to health production and reducing their marginal costs.

    Grossman (1972) study on health demand extends the function of household production for health and assumes that households combine inputs of medical care and time to produce health. Education as an environmental variable increases the efficiency in health production. In this context, the contribution of education to enhanced efficiency in health production has several implications for health care spending. For example, more highly educated individuals and families can economize on their use of medical care, and correspondingly, their spending to produce or restore a given level of health. However, since increased education enhances the returns to healthy days, individuals and families have an incentive to increase their investments in health and potentially, their spending on medical care. Finally, in deciding between inputs of time and medical care in health production, more highly educated households with higher opportunity costs of time and who are likely to have health insurance may seek to economize on inputs of time in health production and as a result, increase their use of medical care and incur additional health care spending.

<h3>II. Dataset Description</h3>

    This study utilizes publicly available data from the Family Income and Expenditure Survey (FIES) as its primary source. This survey was conducted by the Philippine Statistics Authority (PSA) across the country. The FIES serves as the main source of data on family income and expenditure in the Philippines. 

    The initial survey was conducted in 1957 and was subsequently carried out in 1961, 1965, 1971, 1975, and 1979. Since 1985, the FIES has been conducted every three years. Each survey round data collection involves two visits—the first visit captures family income and expenditures for the first half of the year, while the second visit covers data for the second half. For this particular study, the 2015 FIES is used to provide information on the following key variables: household head’s educational attainment, household income, and healthcare expenditure. 

    The FIES has been widely utilized in economic research, serving as a primary data source on human development index, poverty, effects of taxation and social benefits, impact on living standards of social and economic programs, and living standards and disparities across population groups, regions, and time periods.

```python
def plot_stacked_bar_by_region(
    df,
    region_col,
    category_col,
    category_label_map,
    chart_title,
    legend_title="Category",
    colormap="tab20",
    figsize=(14, 7)
):
    import matplotlib.pyplot as plt

    # Group and count
    grouped = df.groupby([region_col, category_col]).size().reset_index(name='count')

    # Map to readable labels
    grouped['CATEGORY_LABEL'] = grouped[category_col].map(category_label_map)

    # Drop any unmapped (NaN) rows
    grouped = grouped.dropna(subset=['CATEGORY_LABEL'])

    # Pivot for stacked bar chart
    pivot_df = grouped.pivot_table(
        index=region_col,
        columns='CATEGORY_LABEL',
        values='count',
        aggfunc='sum'
    ).fillna(0)

    # Plot
    pivot_df.plot(kind='bar', stacked=True, figsize=figsize, colormap=colormap)
    plt.title(chart_title, fontsize=16)
    plt.xlabel("Regions", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.legend(
        title=legend_title,
        title_fontsize=10,
        loc='upper right',
        bbox_to_anchor=(1.25, 1)
    )
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Define labels
hgc_labels = {
    1: "No Grade Completed",
    2: "Preschool",
    3: "Primary",
    4: "Secondary",
    5: "Post-secondary non-tertiary",
    6: "Tertiary",
    7: "Post Baccalaureate"
}

# Plot stacked bar chart
plot_stacked_bar_by_region(
    df=df_2015_v2,
    region_col='W_REGN',
    category_col='HGC_CAT',
    category_label_map=hgc_labels,
    chart_title="Household Head's Educational Attainment by Region in 2015",
    legend_title="Education Level"
)
```

<h2>REGRESSION RESULTS AND DISCUSSION</h2>

    The regression analysis was carried out in five stages, with each subsequent model introducing additional control variables to isolate and better assess the relationship between the household head’s educational attainment and total household health expenditure. The results are presented across Models 1 to 5.

```python
hgc_category_map = {
    "0": 1, "10": 2,
    "210": 3, "220": 3, "230": 3, "240": 3, "250": 3, "260": 3, "280": 3,
    "310": 4, "320": 4, "330": 4, "350": 4,
    "410": 5, "420": 5, "501": 5, "508": 5, "509": 5, "514": 5, "521": 5,
    "522": 5, "531": 5, "532": 5, "534": 5, "542": 5, "544": 5, "548": 5,
    "552": 5, "554": 5, "558": 5, "562": 5, "564": 5, "572": 5, "576": 5,
    "581": 5, "584": 5, "585": 5, "586": 5, "589": 5,
    "601": 6, "614": 6, "621": 6, "622": 6, "631": 6, "632": 6, "634": 6,
    "638": 6, "642": 6, "644": 6, "646": 6, "648": 6, "652": 6, "654": 6,
    "658": 6, "662": 6, "664": 6, "672": 6, "676": 6, "681": 6, "684": 6,
    "685": 6, "686": 6, "689": 6, "810": 6, "820": 6, "830": 6, "840": 6,
    "900": 7
}

# Convert HGC to string, map categories, drop missing
df_2015_v2['HGC'] = df_2015_v2['HGC'].astype(str)
df_2015_v2['HGC_CAT'] = df_2015_v2['HGC'].map(hgc_category_map)
df_2015_v2 = df_2015_v2.dropna(subset=['HGC_CAT'])

# Convert to categorical
df_2015_v2['HGC_CAT'] = df_2015_v2['HGC_CAT'].astype(int).astype('category')
```

```python
df_2015_v2 = df_2015_v2.dropna(subset=['THEALTH', 'TOINC'])
```

```python
# Map 3, 4, and 5 to "3" (Previously Married Category)
df_2015_v2['MS'] = df_2015_v2['MS'].replace({4: 3, 5: 3})
```

```python
categorical_vars = ['HGC_CAT', 'JOB', 'MS', 'SEX', 'HHTYPE', 'W_REGN']

for col in categorical_vars:
    if col in df_2015_v2.columns:
        df_2015_v2[col] = df_2015_v2[col].astype('category')

df_dummies = pd.get_dummies(df_2015_v2, columns=categorical_vars, drop_first=True)
```

```python
# Drop MS_6 if it exists for "Unknown" category in Marital Status
if 'MS_6' in df_dummies.columns:
    df_dummies = df_dummies.drop(columns='MS_6')
```

```python
Y = df_dummies['THEALTH']  # Adjust if column is named differently

# MODEL 1: Education only
X1 = df_dummies.filter(regex='HGC_CAT_')
X1 = sm.add_constant(X1)

# MODEL 2: + Socioeconomic status (includes JOB, PHILHEALTH)
X2 = df_dummies.filter(regex='HGC_CAT_|JOB_')
if 'TOINC' in df_2015_v2.columns:
    X2 = pd.concat([X2, df_2015_v2['TOINC']], axis=1)
X2 = sm.add_constant(X2)

# MODEL 3: + Demographic characteristics
X3 = df_dummies.filter(regex='HGC_CAT_|JOB_|MS_|SEX_')
if 'TOINC' in df_2015_v2.columns and 'AGE' in df_2015_v2.columns:
    X3 = pd.concat([X3, df_2015_v2[['TOINC', 'AGE']]], axis=1)
X3 = sm.add_constant(X3)

# MODEL 4: + Household structure
X4 = df_dummies.filter(regex='HGC_CAT_|JOB_|MS_|SEX_|HHTYPE_')
if 'TOINC' in df_2015_v2.columns and 'FSIZE' in df_2015_v2.columns and 'AGE' in df_2015_v2.columns:
    X4 = pd.concat([X4, df_2015_v2[['TOINC', 'FSIZE', 'AGE']]], axis=1)
X4 = sm.add_constant(X4)

# MODEL 5: + Regional effects
X5 = df_dummies.filter(regex='HGC_CAT_|JOB_|MS_|SEX_|HHTYPE_|W_REGN_')
if 'TOINC' in df_2015_v2.columns and 'FSIZE' in df_2015_v2.columns and 'AGE' in df_2015_v2.columns:
    X5 = pd.concat([X5, df_2015_v2[['TOINC', 'FSIZE', 'AGE']]], axis=1)
X5 = sm.add_constant(X5)
```

```python
def run_ols(X, Y, model_name):
    # Combine and drop rows with any NaNs
    data = pd.concat([X, Y], axis=1).dropna()

    # Separate X and Y
    Y_clean = data[Y.name].astype('float64')
    X_clean = data.drop(columns=[Y.name]).astype('float64')

    # Fit the model
    model = sm.OLS(Y_clean, sm.add_constant(X_clean)).fit()

    print(f"\n=== {model_name} ===")
    print(model.summary())
```

```python
run_ols(X1, Y, "Model 1: Education Only")
run_ols(X2, Y, "Model 2: + Socioeconomic Status")
run_ols(X3, Y, "Model 3: + Demographics")
run_ols(X4, Y, "Model 4: + Household Structure")
run_ols(X5, Y, "Model 5: + Regional Effects")
```

    Model 1 only utilizes the educational attainment categories of the household head (HGC) as explanatory variables. Here we see that only 2.3% of the variance in total household health expenditure is explained by education alone, as indicated by the R-squared value of 0.023. Among the education categories, post-secondary non-tertiary, tertiary, and post-baccalaureate levels are associated with significantly higher household health expenditures, with coefficients increasing in magnitude as the level of education rises. In particular, post-baccalaureate education (HGC_CAT_7) has a strong and statistically significant positive effect on health spending. This result is consistent with theoretical expectations that higher educational attainment enhances health literacy and increases the perceived value of health investment. However, the explanatory power of education alone is limited in the absence of controls for income and other household characteristics.

    Model 2 introduces variables related to socioeconomic status, specifically total household income (TOINC) and employment status of the household head (JOB). The value of R-squared substantially increases to 0.101 with the introduction of the new variables. This suggests that income and employment explain a significant proportion of the variance in total household health expenditure. The addition of the new control variables noticeably reduces the impact of education. For instance, the coefficients for primary, secondary, and post-secondary education are no longer statistically significant as their p-values exceed the 0.05 significance level. Meanwhile, tertiary and post-baccalaureate education levels remain significant, though their effects are smaller. Employment status and total income stand out as strong, positive predictors of health spending, suggesting that income plays a major role in explaining the relationship between education and health expenditure.

    Model 3 introduces variables related to demographic characteristics, specifically marital status (MS), sex (SEX), and age (AGE) of the household head. The addition of these variables increases the value of R-squared slightly to 0.105. The coefficients of the higher education categories, categories 4 to 7, remain positive and statistically significant. In particular, tertiary (HGC_CAT_6) and post-baccalaureate (HGC_CAT_7) show strong significant effects on health expenditure, suggesting that higher educated household heads allocate more spending towards healthcare.

    Additionally, age emerges as a strong predictor of health expenditure, wherein the positive statistically significant coefficient suggests that household health expenditure increases as the age of the household head increases. This may be attributed to greater health needs attributed to age, or increased awareness regarding health. Other demographic variables, such as marital status and sex, do not yield statistically significant effects— wherein they don’t appear to have a meaningful impact on health expenditure once other factors are accounted for.

    Model 4 introduces variables related to household structure, specifically household type (HHTYPE) and family size (FSIZE). The addition of these variables increases the value of R-squared slightly to 0.106. As seen, tertiary and post-baccalaureate education continue to be significantly associated with higher household health expenditure despite showing slightly lower coefficients compared to Model 3. Post-secondary and secondary education also become more stable predictors, though not statistically significant.

    We see that family size has a negative statistically significant coefficient, suggesting that larger households tend to spend less on healthcare— potentially due to budget constraints or competing household needs. Meanwhile, extended families (HHTYPE_2) show significantly higher health spending— potentially due to a larger number of dependents. Additionally, age remains to be a strong positive predictor. Overall, household structure aids in predicting health expenditure, though education, specifically in higher levels, remains to be strongly linked to higher spending. 

    Model 5, the most comprehensive model, introduces region (W_REGN). This model yields the highest R-squared value at 0.108. The addition of the variable of region slightly lowers the coefficients of some categories for education, though the overall finding still holds wherein tertiary and post-baccalaureate education show strong positive significant correlation towards higher total household health expenditure. In addition to this, secondary (HGC_CAT_4) and post-secondary education (HGC_CAT_5) also become statistically significant in this model— further reinforcing that higher education is linked to higher health expenditure.

    Some regional coefficients are statistically significant, wherein households in CALABARZON (W_REGN_41) spend more on healthcare, while those in the National Capital Region (W_REGN_13) spend significantly less—possibly due to better access to public health services. However, most regional effects are not significant, indicating limited geographic disparities once other factors are accounted for. Income, age, and family size remain consistent predictors across all models


<h2>CONCLUSION</h2>

    The results of this study revealed a multidimensional relationship between educational attainment and household health spending. Among all models used, Model 5, with the joint effects of education, income, demographics, household structure, and regional context, proved to be the most robust as it yielded the highest R-squared value. In every version of the regression model tested, moreover, income has consistently exhibited a strong and positive relationship with household health expenditure. Accordingly, higher levels of education, particularly for the ones prepared for categories 4 to 7, showed positive and statistically significant associations with health expenditure. This suggests that more educated households are more capable and inclined to allot a portion of their budget to health. As educational attainment may often lead to better paying jobs, the findings of this study further solidifies and affirms Grossman’s theory that education increases health investment efficiency, and further supports prior studies showing that education has an imperative role in the promotion of healthcare awareness.
    
    The results of this study highlight the multifaceted impact of education on health spending, revealing that it is not isolated as it operates within a broader web of socioeconomic conditions that shape household decisions in complex ways.
